```python
import requests
from bs4 import BeautifulSoup
import datetime 


def news():
    urls = ['https://www.hankyung.com/economy' ,'https://www.hankyung.com/all-news']
    hr =  datetime.datetime.now().hour
    url=urls[1]
    request = requests.get(url)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    # 기사 제목
    links = soup.find_all('h3', 'tit')
    # 기사 내용
    subs = soup.find_all('p', 'read')

    return links, subs
news()
```
